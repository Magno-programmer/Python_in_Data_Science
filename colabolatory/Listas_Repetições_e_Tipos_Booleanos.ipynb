# Boolean

```python
permissoes = []
idades = [5, 18, 22, 1]
```

```python
def verifica_autorizacao_de_direcao(idades, permissoes):
  for idade in idades:
    if idade >= 18:
      permissoes.append(True)
    else:
      permissoes.append(False)

verifica_autorizacao_de_direcao(idades, permissoes)
```

```python
permissoes
```

```python
for permissao in permissoes:
  if permissao == True:
    print('Você ESTA autorizado')
  else:
    print('Você NÃO ESTÁ autorizado')
```
