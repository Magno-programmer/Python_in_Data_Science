# Matplotlib

```python
from random import randrange, seed
import matplotlib.pyplot as plt
```

```python
seed(11)

notas_matematicas = []

for nota in range(8):
  notas_matematicas.append(randrange(0, 11, 2))

notas_matematicas
```

```python
x = list(range(1, 9))
y = notas_matematicas

plt.plot(x, y, marker= 'o')
plt.title('Notas de Matem√°tica')
plt.xlabel('Provas')
plt.ylabel('notas')
plt.yticks(range(0, 11, 2))
plt.grid()
plt.show()
```
